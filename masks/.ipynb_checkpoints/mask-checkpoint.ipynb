<h1>Import modules</h1>

```python
import math ; pi=math.pi
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
import time
import healpy as hp
```

<h1>Dark Energy Survey</h1>

```python
#mask file
file_mask_DES = 'DES-mask-simple-ring-1024.fits'
#Read it
map_mask_DES = hp.read_map(file_mask_DES)
```

```python
#Visualize it
hp.mollview(map_mask_DES)
```

```python
#Compute the power spectrum
LMAX_DES    = 2048
cl_mask_DES = hp.anafast(map_mask_DES, lmax=LMAX_DES)
ell_DES     = np.arange(len(cl_mask_DES))
```

```python
#Plot it
plt.figure(figsize=(10,5))
plt.loglog(ell_DES,cl_mask_DES)
plt.xlabel("$\ell$")
plt.ylabel("$C_{\ell}(M)$")
plt.show()
```

```python
#Write the power spectrum to file
hp.write_cl("Cl-"+file_mask_DES,cl_mask_DES,overwrite=True)
```

```python
#Compute fsky from the mask
#Simplest manner: from the monopole
fsky_monop = np.sqrt(cl_mask_DES[0]/(4*pi))
print(fsky_monop)
#Second manner: from the variance
fsky_sum = ((2*ell_DES+1)/(4*pi)*cl_mask_DES).sum()
print(fsky_sum)
#They give the same result for a binary mask (containing only 0 and 1)
```
